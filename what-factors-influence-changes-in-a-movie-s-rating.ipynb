## Add "Breaking Bad" to the data set 

```python
new_record = {
    'tconst': 'tt0903747',
    'filmtv_id': '1483369',
    'title': 'Breaking Bad',
    'year': 2008,
    'genre': 'Crime',
    'duration': 1200,
    'country': 'United States',
    'directors': "['Michelle MacLaren,Adam Bernstein,Vince Gilligan,Colin Bucksey,Michael Slovis']",
    'actors': "['Bryan Cranston,Aaron Paul,Anna Gunn']",
    'description': "A chemistry teacher diagnosed with inoperable lung cancer turns to manufacturing and selling methamphetamine with a former student in order to secure his family's future.",
    'notes': 'it is the best tvshow you will ever ever see',
    'numVotes': 2000000,
    'averageRating': 9.5,
    'type': 'TV Series',
    'isAdult': 0
}
```

```python
df = df.append(new_record, ignore_index=True)
```

```python
imdb=df.copy()
```

## Drop unimportant columns

```python
imdb.drop(columns=['isAdult','filmtv_id','notes'],inplace=True)
```

## Convert actors & directors cols to lists

```python
import ast

imdb['actors'] = imdb['actors'].fillna('[]')
imdb['directors'] = imdb['directors'].fillna('[]')


convert_to_list = lambda x: ast.literal_eval(x)
converted = lambda x: x[0].split(',') if len(x)!=0 else x
# apply the lambda function to the 'actor_names' column
imdb['actors'] = imdb['actors'].apply(convert_to_list)
imdb['actors'] = imdb['actors'].apply(converted)
imdb['directors'] = imdb['directors'].apply(convert_to_list)
imdb['directors'] = imdb['directors'].apply(converted)
```

```python
imdb
```

## Set weight for each type based on 'numVotes' & 'numVotes'

```python
def calculate_weighted_rating(df):
#      Set the weight threshold (i.e., the minimum number of votes required to be considered)
#     The reason we set the quantile to 0.70 is to ensure that we only consider productions 
#     with a significant number of votes when calculating the weighted rating.
#     For example, if we set the quantile to 0.50, we would be including all productions,
#     even those with very few votes, in our calculation of the weighted rating.
#     This could result in a biased rating, since productions with very few votes 
#     may have a higher or lower rating than they would if they had received more votes
    m = df['numVotes'].quantile(0.70)
   # Calculate the weighted rating and weighted votes for each row
    df['weighted_rating'] = ((df['numVotes'] / (df['numVotes'] + m)) * df['averageRating']) + ((m / (df['numVotes'] + m)) * df['averageRating'].mean())
    return df

# Apply the function to each type of production
imdb = imdb.groupby('type').apply(calculate_weighted_rating)
```
